```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from google.colab.patches import cv2_imshow
from google.colab import files
import requests
from io import BytesIO
```

```python
uploaded = files.upload()
for fn in uploaded.keys():
  image_address = fn
```

```python
img_cv = cv2.imread(image_address)  # Read image
cv2_imshow(img_cv)  # Show image
```

```python
img_cv2 = cv2.cvtColor(np.array(img_cv), cv2.COLOR_RGB2BGR)
img_cv2 = cv2.cvtColor(img_cv2, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_cv2)
```

```python
img_cv2 = cv2.GaussianBlur(img_cv2, (5,5), 0)
denoised = cv2.fastNlMeansDenoising(img_cv2, None, 30, 7, 21)
img_cv2 = cv2.medianBlur(img_cv2, 5)
cv2_imshow(img_cv2)
```
