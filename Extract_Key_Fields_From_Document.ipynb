```python
from google.colab import files
uploaded = files.upload()
for fn in uploaded.keys():
  image_address = fn
```

```python
doc = fitz.open(image_address)
page = doc[0]
```

```python
fields = {
    "Applicant Name": None,
    "Application No": None,
    "Loan Program": None,
    "Interest Rate": None,
    "Term": None,
    "Loan Amount": None,
    "Origination Charges": [],
    "Other Charges": []
}
```

```python
blocks = page.get_text("blocks")

for b in blocks:
    text = b[4].strip()
    bbox = b[:4]

    if "Applicants:" in text:
        fields["Applicant Name"] = {"text": text, "bbox": bbox}
    elif "Application No" in text:
        fields["Application No"] = {"text": text, "bbox": bbox}
    elif "Loan Program" in text or "FIXED" in text or "ARM" in text:
        fields["Loan Program"] = {"text": text, "bbox": bbox}
    elif "Interest Rate" in text:
        fields["Interest Rate"] = {"text": text, "bbox": bbox}
    elif "Term" in text or "mths" in text:
        fields["Term"] = {"text": text, "bbox": bbox}
    elif "Loan Amount" in text or "Total Loan Amount" in text:
        fields["Loan Amount"] = {"text": text, "bbox": bbox}

    if "Underwriting Fee" in text or "Wire Transfer" in text or "Administration Fee" in text:
        fields["Origination Charges"].append({"text": text, "bbox": bbox})
    elif "$" in text and ("Fee" in text or "Charge" in text or "Premium" in text):
        fields["Other Charges"].append({"text": text, "bbox": bbox})
```

```python
print(json.dumps(fields, indent=4))
```
